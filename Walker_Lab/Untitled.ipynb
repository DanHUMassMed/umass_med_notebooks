# Find interesting Genes 

* Here is a small csv with the top 20 up genes from sam-1 with FC and p values.  
* The FCs for let-607 and xbp-1 are there as well.  

* Look at arf-1.1 (#3) as an example:
    * we are interested in where it’s really high in sams-1 (ne_se)
    * not there in let-607 (nl_sl)
    * and xbp-1 is intermediate (nx_sx)  
    
    
* There may be other genes that are shifted in levels, but have similar dynamics.  

```python
import pandas as pd

ne_se_df = pd.read_csv("data/run_NE-ABC_vs_SE-ABC_alldetected.csv",usecols=['ID', 'padj', 'foldChange']) 
ne_se_df = ne_se_df.rename(columns={'padj': 'ne_se_padj', 'foldChange': 'ne_se_foldChange'})
nl_sl_df = pd.read_csv("data/run_NL-ABC_vs_SL-ABC_alldetected.csv",usecols=['ID', 'padj', 'foldChange']) 
nl_sl_df = nl_sl_df.rename(columns={'padj': 'nl_sl_padj', 'foldChange': 'nl_sl_foldChange'})
nx_sx_df = pd.read_csv("data/run_NX-ABC_vs_SX-ABC_alldetected.csv",usecols=['ID', 'padj', 'foldChange'])
nx_sx_df = nx_sx_df.rename(columns={'padj': 'nx_sx_padj', 'foldChange': 'nx_sx_foldChange'})

alldetected_df = ne_se_df.merge(nl_sl_df, on='ID', how='left')
alldetected_df = alldetected_df.merge(nx_sx_df, on='ID', how='left')
alldetected_df
```

```python
ne_se_foldChange_high_df = alldetected_df[alldetected_df['ne_se_foldChange'] > 4]
len(ne_se_foldChange_high_df)
```

```python
nl_sl_foldChange_low = ne_se_foldChange_high_df[ne_se_foldChange_high_df['nl_sl_foldChange'] < 1.8
                                               ]
nl_sl_foldChange_low
```

```python
nx_sx_foldChange_mid = nl_sl_foldChange_low[(nl_sl_foldChange_low['nx_sx_foldChange'] < 2.4) & (nl_sl_foldChange_low['nl_sl_foldChange'] > 1.4)]
print(len(nx_sx_foldChange_mid))
nx_sx_foldChange_mid
```

```python
nx_sx_foldChange_mid.to_csv("data/sweet_spot1.csv",index=False)
```
