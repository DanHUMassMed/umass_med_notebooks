```python
import pandas as pd
```

```python
# Read in the Excel  to get the Sheet Names
xlsx_file_nm = './input_data/WB_phenotypes_AKW_DPH-1.xlsx'
phenotypes_xlsx = pd.ExcelFile(xlsx_file_nm)
```

```python
sheet_names = phenotypes_xlsx.sheet_names
sheet_names
```

```python

one_hot_df = pd.read_excel(xlsx_file_nm, sheet_name='One-Hot_named')
```

```python

one_hot_df.index = one_hot_df['Phenotype']
one_hot_df = one_hot_df.drop('Phenotype', axis=1)
# Fill NaN values with 0 and convert non-NaN values to 1
one_hot_df = one_hot_df.fillna(0).astype(bool).astype(int)
one_hot_df
```

```python
print(f"Not Assigned Count = {one_hot_df['not_assigned'].sum()}")
filtered_df = one_hot_df[one_hot_df['not_assigned'] == 1]
filtered_df
```

# Next Section

```python
wb_phenotypes_df = pd.read_excel(xlsx_file_nm, sheet_name='WB_phenotypes')
unique_values = wb_phenotypes_df['wormbase_id'].unique()
wb_phenotypes_df
```

```python
one_hot_df.columns
slim_phenotypes_list = [name.replace(' ', '') for name in one_hot_df.columns]
slim_phenotype_dict = {phenotype: 0 for phenotype in slim_phenotypes_list}
slim_phenotype_dict['nan'] = 0
slim_phenotype_dict
```

```python
len(one_hot_df.index)
```

```python
def find_slim_phenotype(row_values):
    for index, value in enumerate(row_values):
        if value == 1:
            return slim_phenotypes_list[index]
    return 'not_assigned'
        
phenotype_dict = {'nan':'nan'}
for index, row in one_hot_df.iterrows():
    #print(f"Index: {index}")
    #print(f"Values: {row.values}")
    #print(find_slim_phenotype(row.values))
    phenotype_dict[index] = find_slim_phenotype(row.values)
    #print()

#phenotype_dict
```

```python

wb_id_dict={}
for index, row in wb_phenotypes_df.iterrows():
    #print(f"{row.wormbase_id} {row.phenotype_label})
    if row.wormbase_id not in wb_id_dict:
        wb_id_dict[row.wormbase_id] =slim_phenotype_dict.copy()
        
    column_dict = wb_id_dict[row.wormbase_id]
    #print(f"{row.phenotype_label} {type(row.phenotype_label)}")
    phenotype_label = row.phenotype_label if isinstance(row.phenotype_label, str) else 'nan'
    slim_phenotype = phenotype_dict[phenotype_label]
    column_dict[slim_phenotype] += 1   
    #print(phenotype_dict[row.phenotype_label])
    
wb_id_dict
wb_id_sumamry_df = pd.DataFrame.from_dict(wb_id_dict, orient='index')
wb_id_sumamry_df
```

```python
import shutil
xlsx_file_nm_new = './input_data/WB_phenotypes_AKW_DPH-1.1.xlsx'
shutil.copy(xlsx_file_nm, xlsx_file_nm_new)

with pd.ExcelWriter(xlsx_file_nm_new, mode='a', engine='openpyxl') as writer:
    # Write the DataFrame to a new sheet
    wb_id_sumamry_df.to_excel(writer, sheet_name='wb_id_sumamry')
    
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
df = pd.DataFrame(wb_phenotypes_df, columns=['wormbase_id', 'phenotype_label'])
df
```

```python
# Create a DataFrame from the data


# Create a graph
G = nx.Graph()

# Add edges to the graph
list_of_tuples = list(df.itertuples(index=False, name=None))
G.add_edges_from(list_of_tuples)

# Draw the graph
pos = nx.spring_layout(G, seed=42)  # Seed for reproducibility
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=8, font_weight='bold')
plt.title('Interconnections between wormbase_id and phenotype_label')
plt.show()
```

```python
nx.write_gpickle(G, "graph_file.gpickle")
```

```python
# Draw the graph
pos = nx.spring_layout(G, seed=42)  # Seed for reproducibility
plt.figure(figsize=(12, 24))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=8, font_weight='bold')
plt.title('Interconnections between wormbase_id and phenotype_label')
plt.show()
```

```python
# Convert the graph to a DataFrame
df = pd.DataFrame(G.edges(), columns=['source', 'target'])

# Save the DataFrame to a CSV file
df.to_csv('graph_file.csv', index=False)
```
