```python
import pandas as pd

reference_papers_csv = "./input_data/wormbase_db/wb_reference_papers_ext.csv"
reference_papers_df = pd.read_csv(reference_papers_csv)
reference_papers_df
```

```python
unique_wormbase_ids_df = reference_papers_df['wormbase_id'].value_counts().reset_index()
unique_wormbase_ids_df.columns = ['wormbase_id', 'count']
unique_wormbase_ids_df
```

```python
wormcat_csv = "./input_data/whole_genome_v2_nov-11-2021.csv"
wormcat_df = pd.read_csv(wormcat_csv)
wormcat_df
```

```python
wormcat_w_paper_counts_df = wormcat_df.merge(unique_wormbase_ids_df, left_on="Wormbase ID", right_on="wormbase_id", how="left")
wormcat_w_paper_counts_df['count'] = wormcat_w_paper_counts_df['count'].fillna(0)
wormcat_w_paper_counts_df = wormcat_w_paper_counts_df.rename(columns={'count':'paper_counts'})
wormcat_w_paper_counts_df
```

```python
wormcat_w_paper_counts_cat1_df = wormcat_w_paper_counts_df.groupby('Category 1')['paper_counts'].sum().reset_index()
wormcat_w_paper_counts_cat1_df
```

```python
import matplotlib.pyplot as plt

# Create a pie chart
plt.figure(figsize=(20,20))
plt.pie(wormcat_w_paper_counts_cat1_df['paper_counts'], labels = wormcat_w_paper_counts_cat1_df['Category 1'], autopct='%1.1f%%', pctdistance=0.85)
plt.title('Paper Counts by Category 1')
plt.show()
```

```python
# Reorder the data by 'Category 1' in alphabetical order
wormcat_w_paper_counts_cat1_df = wormcat_w_paper_counts_cat1_df.sort_values(by='paper_counts')

# Create a pie chart
plt.figure(figsize=(20,20))
plt.pie(wormcat_w_paper_counts_cat1_df['paper_counts'], labels = wormcat_w_paper_counts_cat1_df['Category 1'], autopct='%1.1f%%', pctdistance=0.85)
plt.title('% of Papers Published by Wormcat Category 1')

# Adjust label positions to prevent overlap
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a bar chart
plt.figure(figsize=(10,6))
plt.bar(wormcat_w_paper_counts_cat1_df['Category 1'], wormcat_w_paper_counts_cat1_df['paper_counts'])
plt.xlabel('Category 1')
plt.ylabel('Paper Counts')
plt.title('Paper Counts by Wormcat Category 1')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sort the data in descending order
wormcat_w_paper_counts_cat1_df = wormcat_w_paper_counts_cat1_df.sort_values(by='paper_counts', ascending=False)

# Create a bar chart
plt.figure(figsize=(10,6))
sns.barplot(x='Category 1', y='paper_counts', data=wormcat_w_paper_counts_cat1_df)
plt.title('Paper Counts by Wormcat Category 1')
plt.xlabel('Category 1')  # Add x-axis title
plt.ylabel('Paper Counts')  # Change y-axis title
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels

# Add a note to the plot
plt.figtext(0.4, 0.01, 'NOTE: Plot does not consider the number of genes per category', 
            ha='center', fontsize=10, color='gray')


plt.show()
```
