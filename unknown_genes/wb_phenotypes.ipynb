# Create One Hot Encode Sheet
1. Read WB_phenotypes_AKW.xlsx
2. Use the clean_del_phen sheet
3. If Value in any column NaN set to 0 else set to 1
4. create a column "not_assigned" that check if and column has a value if yes set to 1 else set to 0
5. Write Back to Excel with a new sheet One-Hot

```python
import pandas as pd
```

```python
# Read in the Excel Sheet to get the Tab Names
xlsx_file_nm = './input_data/WB_phenotypes_AKW.xlsx'
fpkm_adult_xlsx = pd.ExcelFile(xlsx_file_nm)
```

```python
sheet_names = fpkm_adult_xlsx.sheet_names
sheet_names
```

```python

clean_df = pd.read_excel(xlsx_file_nm, sheet_name='clean_del_phen')
```

```python
clean_df = clean_df.rename(columns={'Unnamed: 0': 'Phenotype'})
clean_df.index = clean_df['Phenotype']
clean_df = clean_df.drop('Phenotype', axis=1)
# Fill NaN values with 0 and convert non-NaN values to 1
clean_df = clean_df.fillna(0).astype(bool).astype(int)
clean_df
```

```python
clean_df['not_assigned'] = (clean_df.iloc[:, :-1].eq(0).all(axis=1)).astype(int)
```

```python
clean_df
```

```python

with pd.ExcelWriter(xlsx_file_nm, mode='a', engine='openpyxl') as writer:
    # Write the DataFrame to a new sheet
    clean_df.to_excel(writer, sheet_name='One-Hot')
```
