```python
import pandas as pd
```

```python
wb_reference_papers = pd.read_csv("./input_data/wormbase_db/wb_reference_papers.csv")
print(len(wb_reference_papers))
print(wb_reference_papers.head())
wb_wbp_id_summary= pd.read_csv("./input_data/wormbase_db/wb_wbp_id_summary.csv")
print(len(wb_wbp_is_summary))
print(wb_wbp_is_summary.head())
```

```python
wb_wbp_id_summary_dict = {}
for index, row in wb_wbp_id_summary.iterrows():
    wb_wbp_id_summary_dict[row['wbp_id']] = row.to_dict()
    
wb_reference_papers['impact_factor']  = float('nan')
wb_reference_papers['pm_id']          = None
wb_reference_papers['pm_issn']        = None
wb_reference_papers['pm_essn']        = None
wb_reference_papers['pm_last_author'] = None

def set_impact_factor(row):
    summamry = wb_wbp_id_summary_dict.get(row['wbp_id'],None)
    if summamry:
        row['impact_factor']  = summamry['impact_factor']
        row['pm_id']          = summamry['pm_id']
        row['pm_issn']        = summamry['pm_issn']
        row['pm_essn']        = summamry['pm_essn']
        row['pm_last_author'] = summamry['pm_last_author']
    return row
    
```

```python
wb_reference_papers = wb_reference_papers.apply(set_impact_factor, axis=1)
wb_reference_papers.head()
```

```python
wb_reference_papers.to_csv('./input_data/wormbase_db/wb_reference_papers_ext.csv',index=False)
```
